```python
def start_pipeline_log(
    system,
    run_id,
    start_time,
    pipeline_name,
    layer_from,
    layer_to,
    table_name
):
    return {
        "system": system,
        "pipeline_name": pipeline_name,
        "layer_from": layer_from,
        "layer_to": layer_to,
        "table_name": table_name,
        "run_id": run_id,
        "start_time": start_time
    }
```

```python
from datetime import datetime

def end_pipeline_log(
    monitoring_table,
    log_dict,
    status,
    record_read,
    record_written,
    dq_issues_count,
    remarks,
    error_message
):

    end_time = datetime.now()

    start_time = log_dict["start_time"]

    # Convert string to datetime if needed
    if isinstance(start_time, str):
        start_time = datetime.fromisoformat(start_time)

    duration = (end_time - start_time).total_seconds()

    final_log = {
        **log_dict,
        "end_time": end_time,
        "duration_seconds": duration,
        "record_read": record_read,
        "record_written": record_written,
        "dq_issues_count": dq_issues_count,
        "status": status,
        "error_message": error_message,
        "remarks": remarks
    }

    df_log = spark.createDataFrame([final_log]) 
    df_log.write.mode("append").option("mergeSchema", "true").saveAsTable(monitoring_table)
```
