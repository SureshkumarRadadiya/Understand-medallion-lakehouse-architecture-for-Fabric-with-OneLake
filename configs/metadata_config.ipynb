```python
df = spark.sql("""
SELECT system,
table_name,
MAX(CASE WHEN key_name = 'type' THEN value END) AS type,
MAX(CASE WHEN key_name = 'fileName' THEN value END) AS fileName,
MAX(CASE WHEN key_name = 'fileFormate' THEN value END) AS fileFormate,
MAX(CASE WHEN key_name = 'delimiter' THEN value END) AS delimiter,
MAX(CASE WHEN key_name = 'rawSchema' THEN value END) AS rawSchema,
MAX(CASE WHEN key_name = 'cleanSchema' THEN value END) AS cleanSchema,
MAX(CASE WHEN key_name = 'loadType' THEN value END) AS loadType,
MAX(CASE WHEN key_name = 'keyColumn' THEN value END) AS keyColumn,
MAX(CASE WHEN key_name = 'watermarkColumn' THEN value END) AS watermarkColumn,
MAX(CASE WHEN key_name = 'watermarkValue' THEN value END) AS watermarkValue,
MAX(CASE WHEN key_name = 'mergeScript' THEN value END) AS mergeScript
FROM config.metadata WHERE tier in ('raw','clean')
GROUP BY table_name,system
""")
```
