```sql
USE [BOAFALakehouse];
IF EXISTS (SELECT 1 FROM BOAFALakehouse.sys.all_object o
                    JOIN BOAFALakehouse.sys.schemas s ON o.schema_id = s.schema_id
                    WHERE o.name = 'view_boafa_transaction' AND s.name = 'clean')
    DROP VIEW clean.view_boafa_transaction;
GO 
    create view clean.[view_boafa_transaction]
    as 
    select
    DUE_DATE,TRANDATE,TRANSACTION_TYPE,ENTITY_ID,LOCATION_ID,
    T.[TRANSACTION_ID],
    TL.TRANSACTION_LINE_ID,
    AMOUNT,GROSS_AMOUNT,AMOUNT_DEBIT,AMOUNT_CREDIT,NET_AMOUNT,
    QUANTITY_COMITTED,QUANTITY_OPEN,AUANTITY_PACKED,QUANTITY_RECEIVED_IN_SHIPMENT,
    ITEM_ID,SUBSIDIARY_ID,TL.CREATE_DATE,
    T.LastModifiedDate,TAL.account
    from clean.[BOAFA_transaction] T
    inner join clean.[BOAFA_transaction] TL
    on T.TRANSACTION_ID=TL.TRASACTION_ID
    left join clean.[BOAFA_transactionline] TAL
    on TL.TRANSACTION_ID=TAL.[transaction] AND TL.TRANSACTION_LINE_ID=TAL.transactionline
```
